```python
#menginport library yang akan digunakan
import pandas as pd
import numpy as np 
```

```python
#membaca dataframe dengan pandas
df = pd.read_csv('Tugas.csv', sep=';')
df.head()
```

```python
#menampilkan info data
df.info()
```

```python
#menampilkan nama kolom
df.columns
```

```python
#menghapus rows data setelah row 1499
df.drop(df.index[1500:13578], inplace = True)
df.head()
```

```python
#casting tipe data 
df['Distance'] = df.Distance.astype(float)
df['YearBuilt'] = df.YearBuilt.str.replace('.00.00', '').astype(float)
df['Landsize'] = df.Landsize.str.replace('.00.00', '').astype(float)
df['BuildingArea'] = df.BuildingArea.str.replace('.00.00', '').astype(float)
df['Postcode'] = df.Postcode.str.replace('.00.00', '')
df['Propertycount'] = df.Propertycount.str.replace('.00.00', '').astype(float)
df
```

```python
#mengubah display data dengan tipe float tanpa angka decimal 
pd.options.display.float_format = '{:,.0f}'.format
def format_float(value):
    return f'{value:,.0f}'
pd.options.display.float_format = format_float
```

```python
#mengisi missing data pada kolom BuildingArea dengan nilai mean
df['BuildingArea'].fillna(df['BuildingArea'].mean(), inplace=True)
```

```python
#mengisi missing data pada kolom YearBuilt dengan nilai mean
df['YearBuilt'].fillna(df['YearBuilt'].mean(), inplace=True)
```

```python
#mengisi missing data pada kolom Councilarea dengan nilai konstanta 0
df['CouncilArea'].fillna(0, inplace=True)
```
