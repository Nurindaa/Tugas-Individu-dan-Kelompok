```python
#convert data string menjadi numeric agar bisa diolah dengan random forest
#Menentukan target adalah TRAG1
num_of_numerical_cols = data_clean._get_numeric_data().columns.shape[0]
print(num_of_numerical_cols, 'numerical columns')
name_of_numerical_cols = data_clean._get_numeric_data().columns
print(name_of_numerical_cols)
```

```python
X = data_clean[['BIO_SEX', 'HISPANIC', 'WHITE', 'BLACK', 'NAMERICAN', 'ASIAN', 'age',
       'ALCEVR1', 'ALCPROBS1', 'marever1', 'cocever1', 'inhever1',
       'cigavail', 'DEP1', 'ESTEEM1', 'VIOL1', 'PASSIST', 'DEVIANT1',
       'SCHCONN1', 'GPA1', 'EXPEL1', 'FAMCONCT', 'PARACTV', 'PARPRES']]
y = data_clean.TREG1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#memodelkan dengan random forest 
#membuat dececio tree sebanyak 20 kali percobaan 
classifier = RandomForestClassifier(n_estimators=20, random_state=0)
classifier.fit(X_train, y_train)

predictions =classifier.predict(X_test)
```

confusion matrix adalah tools untuk evaluasi gambaran prediksi yang sudah dilakukan

```python
confusion_array= sklearn.metrics.confusion_matrix(y_test, predictions)
confusion_array
```

```python
print ('TN= ', confusion_array[0,0] ) #True Negative
print ('FN= ', confusion_array[1,0] ) #False Negative
print ('TP= ', confusion_array[1,1] ) #True Positive
print ('FP= ', confusion_array[0,1] ) #False Positive
```

Baris pertama dari confussion array di atas merupakan non regular smoker prediction : 1440 observation dengan benar di klasifikasi sebagai non regular smoker (called true negatives) and 65 observation dengan salah di klasifikasi sebagai non regular smoker ( false positive). Baris kedua merupakan regular smoker prediction: 234 observation dengan salah diklasifikasi sebagai regular smoker (false negative) dan 91 observation dengan benar diklasifikasi sebagai regular smooker (true positive).
